### Clustering PAM K-medoid (daisy)

```r
gower_dist <- daisy(tordered_data[,c(1:500)],
                    metric = "gower",
                    type = list(logratio = 3))


gower_mat <- as.matrix(gower_dist)

```

```r
# Calculate silhouette width for many k using PAM

sil_width <- c(NA)

for(i in 2:10){
  
  pam_fit <- pam(gower_dist,
                 diss = TRUE,
                 k = i)
  
  sil_width[i] <- pam_fit$silinfo$avg.width
  
}

# Plot sihouette width (higher is better)

plot(1:10, sil_width,
     xlab = "Number of clusters",
     ylab = "Silhouette Width")
lines(1:10, sil_width)
```

```r
pam_fit <- pam(gower_dist, diss = TRUE, k = which(sil_width == max(sil_width, na.rm = T)))
```

```r
pam_fit <- pam(gower_dist, diss = TRUE, k = 2)
```

```r
library(factoextra)
fviz_silhouette(silhouette(pam_fit)) 
```

Une autre fonctionnalité disponible avec pam est une graphique appelée graphique de silhouette. Tout d'abord, une mesure est calculée pour chaque observation afin de déterminer dans quelle mesure elle s'inscrit dans le groupe auquel elle a été affectée. Ceci est effectué en comparant la proximité de l'objet par rapport aux autres objets de son propre cluster avec la proximité des objets dans d'autres clusters. Des valeurs positive indiquent que l’observation est bien placée dans son groupe; des valeurs négative signifient qu'il est probable qu'une observation puisse réellement appartenir à un autre groupe. Au sein de chaque cluster, la valeur de cette mesure est affichée du plus petit au plus grand. Si le tracé de la silhouette montre des valeurs proches de un pour chaque observation, l'ajustement était bon; s'il y a beaucoup d'observations plus proches de zéro, cela indique que l'ajustement n'est pas bon. Le graphique sihouette est très utile pour localiser des groupes dans une analyse par cluster qui peut ne pas bien fonctionner. à leur tour, ces informations peuvent être utilisées pour vous aider à sélectionner le nombre approprié de clusters. Pour l'exemple actuel, voici le tracé de la silhouette pour la solution pam à 2 cluster
Largeur moyenne de la silhouette <0,25 Aucune structure substantielle n'a été trouvée

```r
library(Rtsne)
```

```r
#Visualization in a lower dimensional space

tsne_obj <- Rtsne(gower_dist, is_distance = TRUE,perplexity = 5)

tsne_data <- tsne_obj$Y %>%
  data.frame() %>%
  setNames(c("tsne1", "tsne2")) %>%
  mutate(cluster = factor(pam_fit$clustering),
         name = rownames(tordered_data),
        mutated = factor(ifelse(names(pam_fit$clustering) %in% adverse , "adverse" , ifelse(names(pam_fit$clustering) %in% intermediate, "intermediate", "favorable"))))


MutPlot = ggplot(aes(x = tsne1, y = tsne2), data = tsne_data) +
  geom_point(aes(color = mutated))
print(MutPlot + ggtitle("Risk Group"))


ClustPlot = ggplot(aes(x = tsne1, y = tsne2), data = tsne_data) +
  geom_point(aes(color = cluster))
print(ClustPlot + ggtitle("cluster"))
```

```r
df_IF = AYA_count


samples_IF <-  data.frame(groups = as.character(as.vector(pam_fit$clustering)))

ds_IF <- DESeqDataSetFromMatrix(countData=df_IF, colData=samples_IF, design=~groups)

ds_IF <-DESeq(ds_IF)
res_IF <- results(ds_IF)
```

```r
library("tidyr")
library("dplyr")
library("ggplot2")
```

```r
res <- results(ds_IF, tidy=TRUE) %>%
  arrange(padj, pvalue) %>%
  tbl_df()
head(res)
```

```r
goi <- res$row[1:9]
stopifnot(all(goi %in% names(ds_IF)))
```

```r
tcounts <- t(log2((counts(ds_IF[goi, ], normalized=TRUE, replaced=FALSE)+.5))) %>%
  merge(colData(ds_IF), ., by="row.names") %>%
  gather(gene, expression, (ncol(.)-length(goi)+1):ncol(.))

head(tcounts)
```

```r
library('biomaRt')
mart <- useDataset("hsapiens_gene_ensembl", useMart("ensembl"))
genes <- tcounts$gene
genes <- gsub("\\..*","",genes )
G_list <- getBM(filters= "ensembl_gene_id", attributes= c("ensembl_gene_id",
"hgnc_symbol"),values=genes,mart= mart)
for (i in 1:length(genes)){
    if (genes[i] %in% G_list$ensembl_gene_id){
        a = G_list[G_list$ensembl_gene_id == genes[i],]$hgnc_symbol
        if (a != ""){
            tcounts$gene[i] = a
        }
         
    }
}

ggplot(tcounts, aes(groups, expression, fill=groups)) + 
  geom_boxplot() + 
  facet_wrap(~gene, scales="free_y") + 
  labs(x="Cluster", 
       y="Expression (log normalized counts)", 
       fill="cluster", 
       title="Top 9 Results")
```
